## Step 1: Implement $\epsilon$ annealing

A fairly typical thing to do when using a $\epsilon$-greedy exploration strategy is to anneal $\epsilon$ from its starting value to some final value over a number of timesteps. This allows your agent to explore the environment more at the beginning of training when it knows very little about the environment and then explore less as its policy becomes more and more optimal.

```python
class LinearSchedule(object):                                                                                                                                                                                                                                           
    def __init__(self, schedule_timesteps, final_p, initial_p=1.0):                                                                                                                                                                                                     
        '''
        Linear interpolation between initial_p and final_p over                                                                                                                                                                                                      
        schedule_timesteps. After this many timesteps pass final_p is                                                                                                                                                                                                   
        returned.                                                                                                                                                                                                                                                       
                                                                                                                                                                                                                                                                        
        Args:                                                                                                                                                                                                                                                    
            - schedule_timesteps: Number of timesteps for which to linearly anneal initial_p to final_p                                                                                                                                                                                                                                                  
            - initial_p: initial output value                                                                                                                                                                                                                                        
            -final_p: final output value                                                                                                                                                                                                                                          
        '''                                                                                                                                                                                                                                                       
        self.schedule_timesteps = schedule_timesteps                                                                                                                                                                                                                    
        self.final_p = final_p                                                                                                                                                                                                                                          
        self.initial_p = initial_p                                                                                                                                                                                                                                      
                                                                                                                                                                                                                                                                         
    def value(self, t):                                                                                                                                                                                                                                                 
        fraction = min(float(t) / self.schedule_timesteps, 1.0)                                                                                                                                                                                                         
        return self.initial_p + fraction * (self.final_p - self.initial_p) 
```

## Step 2: Discretize continuous environment

### Discrete Cart Pole

In order to train a tabular agent, such as Sarsa or Q-learning, on a environment with a continuous state space, we need to first discretize the state space. Generally, the finer the discretization the more optimal your policy will be but the longer it will take to train.

For additional info on the Cart-pole problem see [here](https://github.com/openai/gym/wiki/CartPole-v0).

```python
class DiscreteCartPole(object):
    def __init__(self):
        self.env = CartPoleEnv()
        self.action_space = self.env.action_space
        self.round = 1
        
    def reset(self):
        obs = self.env.reset()
        return tuple(np.around(obs, self.round).tolist())
    
    def step(self, action):
        obs, reward, done, info = self.env.step(action)
        return tuple(np.around(obs, self.round).tolist()), reward, done, info
```

# Implement on-policy SARSA

## Step 4: Implement $\epsilon$-Greedy strategy

```python
def egreedy(Q_curr, eps, num_of_actions):
    probs = np.zeros(num_of_actions)
    probs.fill(eps / num_of_actions)
    
    #breaking ties randomly
    q_max = np.max(Q_curr)
    best_action = np.random.choice(np.flatnonzero(Q_curr == q_max))
    
    probs[best_action] = 1 - eps + (eps / num_of_actions)
    return probs
```

## Step 5: Implement SARSA with Linear Schedule for annealing

```python
def sarsa(env, num_episodes, gamma=1.0, alpha=0.1, 
                start_eps=0.2, final_eps=0.1, annealing_steps=1000,
                max_episode_steps=200):
    '''
    Sarsa algorithm.
    
    Args:
        - env: The environment to train the agent on
        - num_episodes: The number of episodes to train the agent for
        - gamma: The discount factor
        - alpha: The stepsize
        - start_eps: The initial epsilon value for e-greedy action selection
        - final_eps: The final epsilon value for the e-greedy action selection
        - annealing_steps: The number of steps to anneal epsilon over
        - max_episode_steps: The maximum number of steps an episode can take
        
    Returns: (Q_func, episode_rewards, episode_lengths)
        - Q: Dictonary mapping state -> action values
        - episode_rewards: Numpy array containing the reward of each episode during training
        - episode_lengths: Numpy array containing the length of each episode during training
    '''
    Q = defaultdict(lambda: np.zeros(env.action_space.n))
 
    episode_rewards = np.zeros(num_episodes)
    episode_lengths = np.zeros(num_episodes)
    
    exploration = LinearSchedule(annealing_steps, final_eps, start_eps)
    summed_episode_step = 0
    
    for x in range(num_episodes):
        # get state
        state = env.reset()
        # start with episode step set to zero
        action_probs = egreedy(Q[state], exploration.value(summed_episode_step), env.action_space.n)
        
        # choose action
        action = np.random.choice(np.arange(env.action_space.n), p=action_probs)
        
        episode_step = 0
        
        # loop for each step of episode
        while episode_step <= max_episode_steps:
            episode_step += 1
            summed_episode_step += 1
            
            # take action and observe next state and reward
            next_state, reward, done, _ = env.step(action)
            
            # book-keeping for plotting
            episode_rewards[x] += reward
            episode_lengths[x] = episode_step
            
            # select next action
            probs = egreedy(Q[next_state], exploration.value(summed_episode_step), env.action_space.n)
            next_action = np.random.choice(np.arange(env.action_space.n), p=probs)
            
            # perform updates
            Q[state][action] += alpha * (reward + gamma * Q[next_state][next_action] - Q[state][action])
            
            state = next_state
            action = next_action
            
            if done:
                break
                
    return Q, episode_rewards, episode_lengths
```

## Step 6: Test implementation on Frozen Lake and CartPole

Testing our implementation on the Frozen-lake (both slippery and non-slippery) environment and the discrete Cart-pole environment. Plot the episode rewards over time averaged over 50 training runs. It helps to smooth this curve over a time window of 100 episodes in order to get  a more clear picture of the learning process. 

```python
def train_sarsa_multiple_run(env, number_of_episodes = 2000, runs = 50, alpha=0.1, final_eps = 0.1):
    '''
    Train SARSA agent for multiple runs
    
    Args:
        - env: environment to run SARSA training on
        - number_of_episodes: number of episodes per run
        - runs: number of runs to be executed
        - alpha: learning rate
        - final_eps: final epsilon value for the Linear Schedule
        
    Returns:
        - mean_rewards: mean reward obtained during the runs
        - mean_lengths: mean length of episodes during the runs        
    '''
    rewards = []
    lengths = []
    
    for x in range(runs):
        _, reward, length = sarsa(env, number_of_episodes, alpha = alpha, final_eps = final_eps)
        
        rewards.append(reward)
        lengths.append(length)
    
    mean_rewards = np.mean(rewards, axis=0)
    mean_lengths = np.mean(lengths, axis=0)
    
    return mean_rewards, mean_lengths
```

```python
def moving_avg(values, window_size=100):
    '''
    Calculate Moving Average
    
    Args:
        - values: values whose moving average has to be calculated
        - window_size: window over which the moving average has to be calculated
    Returns: 
        - moving_avg: moving average over the given window for the given values
    '''
    
    csum = np.cumsum(values)
    csum[window_size:] = csum[window_size:] - csum[:-window_size]
    return csum[window_size-1:] / window_size
```

```python
def plot_reward(rewards, title):
    plt.figure()
    plt.title(title)
    plt.xlabel("Episodes")
    plt.ylabel("Rewards")
    plt.plot(rewards)
```

```python
def plot_length(lengths, title):
    plt.figure()
    plt.title(title)
    plt.xlabel("Episodes")
    plt.ylabel("Lengths")
    plt.plot(lengths)
```

```python
# Non-Slippery Frozen Lake Env
rewards_nsp_frozen_lake, lengths_nsp_frozen_lake = train_sarsa_multiple_run(frozen_lake_env, 1000)

# Calculate moving average of rewards
nsp_frozen_lake_moving_avg_reward = moving_avg(rewards_nsp_frozen_lake)
# Calculate moving average of episode lengths
nsp_frozen_lake_moving_avg_length = moving_avg(lengths_nsp_frozen_lake)

# plot moving average of rewards for non-slippery frozen lake env
plot_reward(nsp_frozen_lake_moving_avg_reward, "Non-Slippery Frozen Lake using SARSA")
# plot moving average of the episode lengths for non-slippery frozen lake env
plot_length(nsp_frozen_lake_moving_avg_length, "Non-Slippery Frozen Lake using SARSA")
```

# Expected SARSA

## Step 7: Implement Expected SARSA

```python
def expected_sarsa(env, num_episodes, gamma=1.0, alpha=0.1, 
                start_eps=0.2, final_eps=0.1, annealing_steps=1000,
                max_episode_steps=200):
    '''
    Expected SARSA algorithm.
    
    Args:
        - env: The environment to train the agent on
        - num_episodes: The number of episodes to train the agent for
        - gamma: The discount factor
        - alpha: The stepsize
        - start_eps: The initial epsilon value for e-greedy action selection
        - final_eps: The final epsilon value for the e-greedy action selection
        - annealing_steps: The number of steps to anneal epsilon over
        - max_episode_steps: The maximum number of steps an episode can take
        
    Returns: (Q_func, episode_rewards, episode_lengths)
        - Q: Dictonary mapping state -> action values
        - episode_rewards: Numpy array containing the reward of each episode during training
        - episode_lengths: Numpy array containing the length of each episode during training
    '''
    Q = defaultdict(lambda: np.zeros(env.action_space.n))
    episode_rewards = np.zeros(num_episodes)
    episode_lengths = np.zeros(num_episodes)
    
    exploration = LinearSchedule(annealing_steps, final_eps, start_eps)
    summed_episode_steps = 0
    
    for x in range(num_episodes):
        state = env.reset()
        episode_steps = 0
        
        while episode_steps <= max_episode_steps:
            episode_steps += 1
            summed_episode_steps += 1
            
            # choose action
            eps = exploration.value(summed_episode_steps)
            probs = egreedy(Q[state], eps, env.action_space.n)
            action = np.random.choice(np.arange(env.action_space.n), p=probs)
            
            # take action
            next_state, reward, done, _ = env.step(action)
            
            # book-keeping for plotting
            episode_rewards[x] += reward
            episode_lengths[x] = episode_steps
            
            # calculate next state expectation for expected SARSA update
            next_state_eps = exploration.value(summed_episode_steps+1)
            next_state_probs = egreedy(Q[next_state], next_state_eps, env.action_space.n)
            
            # update values
            Q[state][action] += alpha * (reward + gamma * np.sum(next_state_probs * Q[next_state]) - Q[state][action])
            
            state = next_state
            
            if done:
                break

    return Q, episode_rewards, episode_lengths
```

## Step 8: Test Expected SARSA implementation

Testing our implentation on the Frozen-lake (both slippery and not slippery) environment and the discrete Cart-pole environment. Plot the episode rewards over time averaged over 50 training runs.
```python
def train_exp_sarsa_multiple_run(env, number_of_episodes = 2000, runs = 50, alpha = 0.1, final_eps = 0.1):
    '''
    Train multiple runs of Expected SARSA
    
    Args:
        - env: environment on which the agent is to be trained
        - number_of_episodes: number of episodes per run
        - runs: number of runs to train the agent
        - alpha: learning rate
        - final_eps: final epsilon value for linear scheduling
    
    Returns:
        - mean_rewards: mean reward obtained during the runs
        - mean_lengths: mean length of episodes during the runs
    '''
    rewards = []
    lengths = []
    
    for x in range(runs):
        _, reward, length = expected_sarsa(env, number_of_episodes,alpha = alpha, final_eps = final_eps)
        
        rewards.append(reward)
        lengths.append(length)
    
    mean_rewards = np.mean(rewards, axis=0)
    mean_lengths = np.mean(lengths, axis=0)
    
    return mean_rewards, mean_lengths
```

# Q-Learning

## Step 9: Implement off-policy Q-Learning

```python
def q_learning(env, num_episodes, gamma=1.0, alpha=0.1, 
                start_eps=0.2, final_eps=0.1, annealing_steps=1000,
                max_episode_steps=200):
    '''
    Q-learning algorithm.
    
    Args:
        - env: The environment to train the agent on
        - num_episodes: The number of episodes to train the agent for
        - gamma: The discount factor
        - alpha: The stepsize
        - start_eps: The initial epsilon value for e-greedy action selection
        - final_eps: The final epsilon value for the e-greedy action selection
        - annealing_steps: The number of steps to anneal epsilon over
        - max_episode_steps: The maximum number of steps an episode can take
        
    Returns: (Q_func, episode_rewards, episode_lengths)
        - Q: Dictonary mapping state -> action values
        - episode_rewards: Numpy array containing the reward of each episode during training
        - episode_lengths: Numpy array containing the length of each episode during training
    '''
    Q = defaultdict(lambda: np.zeros(env.action_space.n))
    episode_rewards = np.zeros(num_episodes)
    episode_lengths = np.zeros(num_episodes)
    summed_episode_steps = 0
    
    exploration = LinearSchedule(annealing_steps, final_eps, start_eps)
    
    for x in range(num_episodes):
        state = env.reset()
        episode_step = 0
        
        while episode_step <= max_episode_steps:
            episode_step += 1
            summed_episode_steps += 1
            
            # pick action
            epsilon = exploration.value(summed_episode_steps)
            probs = egreedy(Q[state], epsilon, env.action_space.n)
            action = np.random.choice(np.arange(env.action_space.n), p=probs)
            
            # take action and observe reward and next state
            next_state, reward, done, _ = env.step(action)
            
            # book-keeping for plotting
            episode_rewards[x] += reward
            episode_lengths[x] = episode_step
            
            # perform updates
            max_next_state_q = np.max(Q[next_state])
            Q[state][action] += alpha * (reward + gamma * max_next_state_q - Q[state][action])
            
            state = next_state
            
            if done:
                break
            
    return Q, episode_rewards, episode_lengths
```

## Step 10: Test our Q-Learning implementation

Testing our implentation on the Frozen-lake (both slippery and not slippery) environment and the discrete Cart-pole environment. Plot the episode rewards over time averaged over 50 training runs. It helps to smooth this curve over a time window of 100 episodes in order to get a more clear picture of the learning process.

```python
def train_q_learning_multiple_run(env, num_episodes = 500, runs = 50, alpha=0.1, final_eps=0.1):
    '''
    Train multiple runs of Q-Learning
    
    Args:
        - env: environment on which the agent is to be trained
        - number_of_episodes: number of episodes per run
        - runs: number of runs to train the agent
        - alpha: learning rate
        - final_eps: final epsilon value for linear scheduling
    
    Returns:
        - mean_rewards: mean reward obtained during the runs
        - mean_lengths: mean length of episodes during the runs
    '''
    rewards = []
    lengths = []
    
    for _ in range(runs):
        _, reward, length = q_learning(env, num_episodes, alpha = alpha,final_eps = final_eps)
        rewards.append(reward)
        lengths.append(length)
        
    mean_rewards = np.mean(rewards, axis=0)
    mean_lengths = np.mean(lengths, axis=0)
    
    return mean_rewards, mean_lengths
```
